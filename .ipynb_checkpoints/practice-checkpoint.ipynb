# Load the book

```python
with open('miracle_in_the_andes.txt', 'r') as file:
    book = file.read()
```

# Extract the paragraphs where 'love' was used

```python
from re import compile, findall

love_pattern = compile('[^\n]+love[^\n]+')
love_query = findall(love_pattern, book)
love_query
```

# Extract the chapter titles

```python
chapter_pattern = compile('([A-Za-z ]+)\n\n')
# () tells compile to return only the matching pattern in ()
chapter_query = findall(chapter_pattern, book)
chapter_query
```

# Function that finds the occurrence of any word

```python
from collections import Counter

def find(target_word):
    word_pattern = compile('[A-Za-z]+')
    word_query = findall(word_pattern, book.lower())
    
    # word_freq_map = Counter(word_query)
    # not sure why spliting by space gives different result
    # word_freq_map = Counter(book.lower().split())
    
    word_freq_map = {}
    
    for single_word in word_query:
        word_freq_map[single_word] = word_freq_map.get(single_word, 0) + 1
    
    
    try:
        return word_freq_map[target_word]
    except:
        return f'The book does not contain the word "{target_word}"'
    
    # Counter returns 0 for keys that doesn't exist, so we have to write code this way
    # return word_freq_map[target_word] if target_word in word_freq_map else f'The book does not contain the word "{target_word}"'
```

# Call the function

```python
find('love')
```
